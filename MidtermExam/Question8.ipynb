```python
#IMPROVED CODE
class Dog:
    """
    Represents a single dog with name, age, and breed.
    """
    def __init__(self, name: str, age: int, breed: str):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self) -> None:
        """Simulate the dog barking."""
        print(f"{self.name} says: Woof! ")


class Kennel:
    """
    Represents a kennel that can contain multiple Dog objects.
    Demonstrates composition (Kennel *has many* Dogs).
    """
    def __init__(self):
        self.dogs = []

    def add_dog(self, dog: Dog) -> None:
        """Add a new Dog to the kennel."""
        self.dogs.append(dog)
        print(f"Added {dog.name} ({dog.breed}, {dog.age} yrs old) to the kennel.")

    def remove_dog(self, name: str) -> bool:
        """Remove a Dog by name. Returns True if removed, False if not found."""
        for d in self.dogs:
            if d.name.lower() == name.lower():
                self.dogs.remove(d)
                print(f"Removed {name} from the kennel.")
                return True
        print(f"No dog named {name} found.")
        return False

    def oldest_dog(self) -> Dog | None:
        """Return the oldest Dog in the kennel, or None if empty."""
        if not self.dogs:
            return None
        return max(self.dogs, key=lambda d: d.age)

    def find_by_breed(self, breed: str) -> list[Dog]:
        """Return a list of Dogs of the given breed."""
        return [d for d in self.dogs if d.breed.lower() == breed.lower()]


# --- Minimal tests ---
if __name__ == "__main__":
    # Create some dogs
    buddy = Dog("Buddy", 5, "Labrador")
    maxy = Dog("Max", 3, "Beagle")
    luna = Dog("Luna", 7, "Labrador")

    # Create a kennel and add dogs
    kennel = Kennel()
    kennel.add_dog(buddy)
    kennel.add_dog(maxy)
    kennel.add_dog(luna)

    print("\n All dogs in kennel:")
    for d in kennel.dogs:
        print(f"- {d.name} ({d.age} yrs, {d.breed})")

    print("\nOldest dog:")
    oldest = kennel.oldest_dog()
    if oldest:
        print(f"{oldest.name} is the oldest at {oldest.age} years old.")

    print("\nFind all Labradors:")
    labs = kennel.find_by_breed("Labrador")
    for d in labs:
        print(f"- {d.name} ({d.age} yrs)")

    print("\nRemoving Max...")
    kennel.remove_dog("Max")

    print("\nRemaining dogs:")
    for d in kennel.dogs:
        print(f"- {d.name}")

    print("\nLet one dog bark:")
    buddy.bark()



```

```python
#CHATGPT CODE
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        return "woof"


class Kennel:
    def __init__(self):
        self.dogs = []

    def add_dog(self, dog):
        self.dogs.append(dog)

    def remove_dog(self, name):
        self.dogs = [d for d in self.dogs if d.name != name]

    def oldest_dog(self):
        return max(self.dogs, key=lambda d: d.age)

    def find_by_breed(self, breed):
        return [d for d in self.dogs if d.breed == breed]


# ---- Function calls / tests ----
if __name__ == "__main__":
    # Create some Dog objects
    dog1 = Dog("Buddy", 5, "Labrador")
    dog2 = Dog("Max", 3, "Beagle")
    dog3 = Dog("Luna", 7, "Labrador")

    # Create a Kennel and add the dogs
    kennel = Kennel()
    kennel.add_dog(dog1)
    kennel.add_dog(dog2)
    kennel.add_dog(dog3)

    # Display all dogs
    print("All dogs in the kennel:")
    for d in kennel.dogs:
        print(f"- {d.name} ({d.age} yrs, {d.breed})")

    # Show the oldest dog
    oldest = kennel.oldest_dog()
    print(f"\nOldest dog: {oldest.name}, Age: {oldest.age}")

    # Find all Labradors
    print("\nLabradors in the kennel:")
    for d in kennel.find_by_breed("Labrador"):
        print(f"- {d.name}")

    # Remove one dog
    print("\nRemoving Max...")
    kennel.remove_dog("Max")

    print("\nRemaining dogs:")
    for d in kennel.dogs:
        print(f"- {d.name}")

    # Let a dog bark
    print(f"\n{dog1.name} says: {dog1.bark()}")

```
