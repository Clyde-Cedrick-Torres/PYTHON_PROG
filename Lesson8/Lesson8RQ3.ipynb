```python
from pathlib import Path
import shutil

# Step 1: Create  folder
base = Path.home() / "my_folder"
base.mkdir(exist_ok=True)
print(f"Created directory: {base}")
```

```python
# Step 2: Create three files inside my_folder
(base / "file1.txt").write_text("File 1 content")
print("Created file: file1.txt")

(base / "file2.txt").write_text("File 2 content")
print("Created file: file2.txt")

(base / "image1.png").touch()
print("Created file: image1.png")
```

```python
#Step 3: Create an 'images' subfolder and move image1.png
images_dir = base / "images"
images_dir.mkdir(exist_ok=True)
shutil.move(str(base / "image1.png"), str(images_dir / "image1.png"))
print("Moved image1.png to images/ folder")
```

```python
# Step 4: Delete file1.txt
(base / "file1.txt").unlink(missing_ok=True)
print("Deleted file: file1.txt")
```

```python
# Step 5: Delete the entire my_folder directory 
shutil.rmtree(base)
print("Deleted the entire my_folder directory and its contents.")
```

```python
from pathlib import Path
import shutil

practice = Path.home() / "practice_files"
images_dir = practice / "images"
images_dir.mkdir(parents=True, exist_ok=True)

for img in practice.glob("*.png"):
    shutil.move(str(img), str(images_dir / img.name))

for img in practice.glob("*.jpg"):
    shutil.move(str(img), str(images_dir / img.name))

for img in practice.glob("*.gif"):
    shutil.move(str(img), str(images_dir / img.name))

print("All image files moved to images/ folder!")
```
